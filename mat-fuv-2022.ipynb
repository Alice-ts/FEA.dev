```python
# Bibliotecas usadas
import math
import numpy as np

# Iniciando o processo, vamos definir nossas variáveis
AB = float(input("Lado oposto ao ângulo (AB): "))
OB = float(input("Outro lado dado (ÔB): "))
lado = float(input("Lado do quadrado inscrito: "))

# Aguardando
print('Aguarde...')

# Aqui é onde a mágica acontece! Essa é a lei dos cossenos, aplicada milhões de vezes
acosines = [math.acos((OB**2+z**2-AB**2)/(2*OB*z)) for z in np.arange(lado**1/2, lado, 0.00001)]

# Como queremos o cosseno do maior arco cosseno, basta dizer isso para o Python (e um round para arredondar)
answer = round(math.cos(max(acosines)), 5)

# Agora, veremos as respostas
print(f'Resposta em decimal:', answer)
print(f'Resposta em (numerador, denominador): ', (answer).as_integer_ratio())
```

```python
# Bibliotecas usadas
from sympy import symbols, Eq, solve
from numpy.polynomial import polynomial as P

# Dando nome às variáveis da nossa equação
a, b = symbols('a b')

# Definindo as equações da questão
eq1 = Eq(2*a + b, 0)
eq2 = Eq(a**2*b - 2, 0)

# A função solve resolve a equeção para você!!!!
sol = solve((eq1, eq2))
roots = sol[0]

# Vamos dar uma olhada?
print(f'dicionário das raízes:', roots)

# Agora, a função polyfromroots nos dá os coeficienetes do polinômiko que possuias raízes dadas, mas em ordem inversa!
coef = P.polyfromroots((roots[a], roots[a], roots[b]))

# O m está na segunda posição da lista (note que o Python conta começando do 0)
coefm = coef[1]

# Agora, basta somar o coeficente com a raíz encontrada
answer = coefm + roots[a]

# Vamos dar uma olhada em como ficou a resposta!
print(f'resposta:', answer)
```
